# Python Basics: Basic Arithmetic, Lists, Indexes, and Their Manipulation

```python
# Simple computation can be done in Python without imports
print(1 + 1) # basic arithmetic
print("a" + "a") # string concatenation
print(True + False) # Boolean logic

print("Be sure '" + str(2) + "' have the correct type() when computing strings and numbers!")
```

```python
# Lists can also be created without imports
list1 = [1, "a", True] # simple list
list2 = [["x" , "y"], "z"] # composite list
list3 = [2 + 2, 2 * 2 + 1, 2 ** 3 - 2] # list with computation
the_list = [list1, list2, list3] # compositing lists

print(list1)
print(list2)
print(list3)
print(the_list)
```

```python
# Access elements of a list or lists of a list using index notation
print(list1[0]) # print first element of list1
print(list1[-1]) # print last element of list1
print(list3[:2]) # print first two elements
print(list3[0:2]) # this is the same as above: print the first two elements
print(list1[-2:]) # print the last two elements

print(list2[0][1]) # print the first list and the second element of the first list
print(the_list[1][0][1]) #print the first list of the second list with the second element
```

```python
# List manipulation and index usage
a_long_list = [list3[0], list3[1], list3[2], list2[0][0], list2[0][1], list2[1]]
a_different_list = a_long_list[:] # copies all of a_long_list to preserve memory reference to a_long_list; otherwise a_different_list = a_long_list share memory reference
a_different_list[0] = 4.1; a_different_list[3] = "ex"; del(a_different_list[-1]) # altering a_different_list; notice the semi-colon, which acts as newline for executing code
a_shorter_list = a_long_list[:]; del(a_shorter_list[-4:-1]) # creating a copy of a_long_list and shortening the list
a_big_list = a_long_list + a_different_list + a_shorter_list # simple list arithmetic (only addition)

print(a_long_list) # making a list with list elements
print(a_different_list) # altering a_long_list
print(a_long_list[-4:-1]) # notice the order of the negative index; the second index is excluded, which means -4 through -2 are recalled
print(a_shorter_list) # removing middle elements from a_long_list
print(a_big_list) # adding all lists together
```
