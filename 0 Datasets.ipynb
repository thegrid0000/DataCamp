```python
#https://scikit-learn.org/stable/datasets/
from sklearn.datasets import load_boston, load_iris, load_diabetes, load_digits, load_linnerud, load_wine, load_breast_cancer
import pandas as pd

#Load and return the boston house-prices dataset (regression).
boston = load_boston() 
pd_boston = pd.DataFrame(data=boston['data'], columns=boston['feature_names'])
pd_boston_tar = pd.DataFrame(data=boston['target'])
#Load and return the iris dataset (classification).
iris = load_iris() 
pd_iris = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
pd_iris_tar = pd.DataFrame(data=iris['target'])
#Load and return the diabetes dataset (regression).
diabetes = load_diabetes() 
pd_diabetes = pd.DataFrame(data=diabetes['data'], columns=diabetes['feature_names'])
pd_diabetes_tar = pd.DataFrame(data=diabetes['target'])
#Load and return the digits dataset (classification).
digits = load_digits() 
#Load and return the linnerud dataset (multivariate regression).
linnerud = load_linnerud() 
pd_linnerud = pd.DataFrame(data=linnerud['data'], columns=linnerud['feature_names'])
pd_linnerud_tar = pd.DataFrame(data=linnerud['target'], columns=linnerud['target_names'])
#Load and return the wine dataset (classification).
wine = load_wine() 
pd_wine = pd.DataFrame(data=wine['data'], columns=wine['feature_names'])
pd_wine_tar = pd.DataFrame(data=wine['target'])
#Load and return the breast cancer wisconsin dataset (classification).
breast_cancer = load_breast_cancer()
pd_breast_cancer = pd.DataFrame(data=breast_cancer['data'], columns=breast_cancer['feature_names'])
pd_breast_cancer_tar = pd.DataFrame(data=breast_cancer['target'])
```
